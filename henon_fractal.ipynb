```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Henon map parameters
a = 1.4
b = 0.3
num_points = 100000  # Use more points to see finer structure
x, y = 0.1, 0.1

def generate_henon(a, b, num_points=100000):
    x_vals, y_vals = [], []
    x, y = 0.1, 0.1
    for _ in range(num_points):
        x_new = 1 - a * x**2 + y
        y_new = b * x
        x, y = x_new, y_new
        x_vals.append(x)
        y_vals.append(y)
    return x_vals, y_vals

x_vals, y_vals = generate_henon(a, b)

def plot_henon(a, b, xlim_min, xlim_max, ylim_min, ylim_max):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.scatter(x_vals, y_vals, s=0.2, color='black', alpha=0.6)
    ax.set_xlim(xlim_min, xlim_max)
    ax.set_ylim(ylim_min, ylim_max)
    ax.set_title(f'HÃ©non Map Attractor')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True, linestyle='--', alpha=0.3)
    plt.show()

a_slider = widgets.FloatSlider(value=1.4, min=1.0, max=1.45, step=0.01, description='a:')
b_slider = widgets.FloatSlider(value=0.3, min=0.1, max=0.4, step=0.01, description='b:')
xlim_min_slider = widgets.FloatSlider(value=-1, min=-2, max=0, step=0.1, description='x min:')
xlim_max_slider = widgets.FloatSlider(value=1, min=0, max=2, step=0.1, description='x max:')
ylim_min_slider = widgets.FloatSlider(value=-1, min=-2, max=0, step=0.1, description='y min:')
ylim_max_slider = widgets.FloatSlider(value=1, min=0, max=2, step=0.1, description='y max:')

def update_plot(change):
    clear_output(wait=True)  
    display(ui)  
    plot_henon(a_slider.value, b_slider.value, xlim_min_slider.value, xlim_max_slider.value, ylim_min_slider.value, ylim_max_slider.value)

a_slider.observe(update_plot, names='value')
b_slider.observe(update_plot, names='value')
xlim_min_slider.observe(update_plot, names='value')
xlim_max_slider.observe(update_plot, names='value')
ylim_min_slider.observe(update_plot, names='value')
ylim_max_slider.observe(update_plot, names='value')

ui = widgets.VBox([
    xlim_min_slider, xlim_max_slider, 
    ylim_min_slider, ylim_max_slider
])

display(ui)
plot_henon(a_slider.value, b_slider.value, xlim_min_slider.value, xlim_max_slider.value, ylim_min_slider.value, ylim_max_slider.value)
```
